```python
import json
import re
import requests
from BeautifulSoup import *
```

```python
url="https://www.youtube.com/feed/trending"
response = requests.get(url)
page= response.text
soup=BeautifulSoup(page)
tags=soup.findAll('a', attrs={'class':"yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link "})
duration = soup.findAll('span', attrs={'class': 'video-time'})
user = soup.findAll('a', attrs={'class': 'g-hovercard yt-uix-sessionlink       spf-link '})
data = soup.findAll("ul", attrs={ 'class':"yt-lockup-meta-info", } ) 
new=str(data)
views=re.findall('>([0-9]+\S*)\s+views',new)
first_list=[]
for i,j,z,x in zip(tags,duration,user,views):
    first_list.append({'Title' :i.get("title"),"URL" : "https://www.youtube.com"+i.get('href'),'Duration':j.text,"Username":z.text,"Views": x})
print(first_list)

with open('trending.json', 'w') as output:
    json.dump(first_list, output, sort_keys = True, indent = 4)

import unicodecsv as csv

with open('trending__.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Title", "URL", "Username", "Duration", "Views"])
    for i in first_list:
        writer.writerow([i['Title'],i['URL'],i['Username'],i['Duration'],i['Views']])
```
